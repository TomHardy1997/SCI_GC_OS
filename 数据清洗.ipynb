```python
import pandas as pd

# 读取数据
tcga = pd.read_table('/home/stat-jijianxin/project/gc_code/stad_tcga_pan_can_atlas_2018/data_clinical_patient.txt')

# 提取生存信息相关的列
survival_columns = [
    '#Patient Identifier', 
    'Overall Survival Status', 
    'Overall Survival (Months)',
    'Disease-specific Survival status',
    'Months of disease-specific survival',
    'Progression Free Status',
    'Progress Free Survival (Months)'
]

# 选取包含上述列的子集
survival_df = tcga[survival_columns]

# 将第三行作为列名
new_columns = survival_df.iloc[3]
survival_df.columns = new_columns

# 移除前三行并重置索引
survival_df = survival_df.iloc[4:].reset_index(drop=True)

# 确认列名是否正确
print(survival_df.columns)

# 将 OS_MONTHS 列转换为数值类型
survival_df['OS_MONTHS'] = pd.to_numeric(survival_df['OS_MONTHS'], errors='coerce')

# 选择 OS_MONTHS >= 1 的行
filtered_df = survival_df[survival_df['OS_MONTHS'] >= 1]

# 显示结果
print(filtered_df)
```
